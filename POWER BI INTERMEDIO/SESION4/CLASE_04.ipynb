```
CANTIDADES VENDIDAS X MES = 
ADDMISSINGITEMS
(
    SUMMARIZE
    (
        Factura,
        Factura[Ao],
        Factura[Mes],
        "Cantidad",SUMX(factura,Factura[cnt_ven_fac_in]-Factura[cnt_ret_fac_in])
    )
)
```





```
CANTIDADES VENDIDAS X CATEGORA = 
GROUPBY(
    Factura,
    Factura[Categora],
    "Cantidad",SUMX(CURRENTGROUP(),Factura[cnt_ven_fac_in]-Factura[cnt_ret_fac_in])
)
```
