# **La tabla CALENDARIO**

## **Primera forma CALENDARAUTO:**

La funcin CALENDARAUTO retorna una tabla de una sola columna con las fechas consecutivas desde el 1 de enero del ao mas pequeo detectado automticamente en el modelo hasta el 31 de diciembre del ao ms grande detectado automticamente en el modelo.



```
CALENDARIO = CALENDARAUTO()
```



## **Segunda forma CALENDAR:**

La funcin CALENDAR retorna una tabla de una sola columna con la fechas consecutivas dadas desde una fecha inicial hasta una fecha final.

### **Sintaxis**



```
CALENDAR( <Fecha Inicial>, <Fecha Final> )
```



### **Ejemplo 1:**



```
CALENDARIO = 
CALENDAR
(
    MIN(VENTA[OrderDate]),
    MAX(VENTA[OrderDate])
)
```



### **Ejemplo 2:**



```
CALENDARIO = 
CALENDAR
(
    FIRSTDATE(VENTA[OrderDate]),
    LASTDATE(VENTA[OrderDate])
)
```



## **Tercera forma CALENDAR usando variables y funciones DAX:**



```
CALENDARIO = 
VAR Fecha =
CALENDAR
(
    FIRSTDATE(VENTA[OrderDate]),
    LASTDATE(VENTA[OrderDate])
)
RETURN
SELECTCOLUMNS
(
    Fecha,
    "Fecha", [Date],
    "Ao", YEAR([Date]),
    "Semestre", "S-" & ROUNDUP(MONTH([Date])/6, 0),
    "Trimestre", "T-" & ROUNDUP(MONTH([Date])/3, 0),
    "Trim", ROUNDUP(MONTH([Date])/3, 0),
    "Mes", UPPER(FORMAT([Date],"mmmm")),
    "MesNum", MONTH([Date]),
    "Da", UPPER(FORMAT([Date],"dddd")),
    "DiaNum", WEEKDAY([Date])
)
```
