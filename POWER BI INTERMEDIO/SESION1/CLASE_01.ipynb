# **Mdulo 1: El modelado de los datos**

El **modelado de datos en Power BI** es una de las partes ms importantes del proceso de creacin de informes y visualizaciones efectivas.

El modelado de datos permite a los usuarios crear relaciones entre diferentes conjuntos de datos, definir jerarquas, crear campos calculados, medidas y tablas calculadas.

Estas funciones permiten a los usuarios analizar los datos de manera ms efectiva y obtener informacin valiosa.

Al modelar los datos de manera adecuada, los usuarios pueden crear informes y visualizaciones que proporcionen informacin clara y concisa sobre el rendimiento de la empresa, la rentabilidad de los productos, la eficiencia de los procesos, entre otros.

El modelado de datos permite a los usuarios mejorar la eficiencia en el procesamiento de datos, lo que se traduce en una mayor velocidad y capacidad para procesar grandes cantidades de informacin.

Tambin permite a los usuarios estructurar la informacin de manera ms intuitiva y fcil de entender para el usuario final.



Aqu hay algunas razones por las que el modelado de datos es importante en Power BI:

1. **Precisin en los resultados**: El modelado de datos garantiza que los datos sean precisos y coherentes en todos los informes y anlisis. Al establecer relaciones correctas entre las tablas, se pueden realizar clculos precisos en los informes.

2. **Mayor eficiencia**: El modelado de datos permite que los informes se ejecuten ms rpido y sean ms eficientes. Al utilizar medidas y columnas calculadas en lugar de frmulas repetitivas en los informes, se ahorra tiempo y se reduce el riesgo de errores.

3. **Flexibilidad**: El modelado de datos permite una mayor flexibilidad en la creacin de informes y anlisis. Al establecer relaciones entre las tablas de datos, se pueden crear visualizaciones y grficos personalizados y relacionados.

4. **Escalabilidad:** El modelado de datos tambin permite una mayor escalabilidad en los informes y anlisis. Al tener un modelo de datos slido, se pueden agregar fcilmente nuevas tablas de datos y visualizaciones sin afectar el rendimiento del informe.

## **2.- Concepto de tabla**

Una tabla es una estructura de datos que contiene informacin organizada en filas y columnas.

Las tablas son una parte fundamental del modelo de datos en Power BI, ya que se utilizan para almacenar y relacionar informacin de varias fuentes de datos.

## **3.- Concepto de clave primaria (PK)**

Una clave primaria es un atributo o conjunto de atributos en una tabla de base de datos que identifica de manera nica cada fila en esa tabla.

La clave primaria se utiliza para garantizar que no haya filas duplicadas en la tabla y que cada fila tenga una identificacin nica.

En una tabla, la clave primaria se define mediante la seleccin de una o ms columnas que tienen valores nicos en cada fila.

Estas columnas se definen como clave primaria y se utiliza para relacionar datos entre diferentes tablas en una base de datos.

Las claves primarias son importantes porque ayudan a garantizar la integridad de los datos en una base de datos.

Al garantizar que cada fila tenga una identificacin nica, se evita la posibilidad de datos duplicados o informacin incompleta.

## **4.- Concepto de clave externa o clave fornea (FK)**

Una **clave externa (FK)** es una columna o varias columnas en una tabla que se relaciona con una columna o varias columnas en otra tabla.

La columna que se relaciona en la otra tabla es la clave principal.

La relacin entre las dos tablas se establece cuando se igualan los valores de la clave externa con los valores de la clave principal.

# **Funcin CALCULATE**:

En Power BI, la funcin CALCULATE se utiliza para modificar el contexto de evaluacin de una frmula o expresin. 

Permite agregar o eliminar filtros en una **columna calculada** o **medida**, lo que puede cambiar los resultados de las visualizaciones o informes.

La funcin CALCULATE puede ser utilizada junto con otras funciones de agregacin, como **SUM**, AVERAGE, MIN, **MAX**, entre otras, para calcular valores en un contexto especfico.

Por ejemplo, se puede utilizar la funcin CALCULATE para calcular la suma de ventas de un producto slo para una regin especfica o para un perodo de tiempo determinado.

## **Sintaxis:**



```
CALCULATE(<expresin a calcular>, <filtros>)
```



Donde <expresin a calcular> es la medida o columna que se desea calcular y <filtros> son los filtros que se desean aplicar a la expresin.

Los filtros pueden ser especificados utilizando funciones como FILTER, ALL, VALUES, entre otras.
