# **Módulo 1: El modelado de los datos**

El **modelado de datos en Power BI** es una de las partes más importantes del proceso de creación de informes y visualizaciones efectivas.

El modelado de datos permite a los usuarios crear relaciones entre diferentes conjuntos de datos, definir jerarquías, crear campos calculados, medidas y tablas calculadas.

Estas funciones permiten a los usuarios analizar los datos de manera más efectiva y obtener información valiosa.

Al modelar los datos de manera adecuada, los usuarios pueden crear informes y visualizaciones que proporcionen información clara y concisa sobre el rendimiento de la empresa, la rentabilidad de los productos, la eficiencia de los procesos, entre otros.

El modelado de datos permite a los usuarios mejorar la eficiencia en el procesamiento de datos, lo que se traduce en una mayor velocidad y capacidad para procesar grandes cantidades de información.

También permite a los usuarios estructurar la información de manera más intuitiva y fácil de entender para el usuario final.



Aquí hay algunas razones por las que el modelado de datos es importante en Power BI:

1. **Precisión en los resultados**: El modelado de datos garantiza que los datos sean precisos y coherentes en todos los informes y análisis. Al establecer relaciones correctas entre las tablas, se pueden realizar cálculos precisos en los informes.

2. **Mayor eficiencia**: El modelado de datos permite que los informes se ejecuten más rápido y sean más eficientes. Al utilizar medidas y columnas calculadas en lugar de fórmulas repetitivas en los informes, se ahorra tiempo y se reduce el riesgo de errores.

3. **Flexibilidad**: El modelado de datos permite una mayor flexibilidad en la creación de informes y análisis. Al establecer relaciones entre las tablas de datos, se pueden crear visualizaciones y gráficos personalizados y relacionados.

4. **Escalabilidad:** El modelado de datos también permite una mayor escalabilidad en los informes y análisis. Al tener un modelo de datos sólido, se pueden agregar fácilmente nuevas tablas de datos y visualizaciones sin afectar el rendimiento del informe.

## **2.- Concepto de tabla**

Una tabla es una estructura de datos que contiene información organizada en filas y columnas.

Las tablas son una parte fundamental del modelo de datos en Power BI, ya que se utilizan para almacenar y relacionar información de varias fuentes de datos.

## **3.- Concepto de clave primaria (PK)**

Una clave primaria es un atributo o conjunto de atributos en una tabla de base de datos que identifica de manera única cada fila en esa tabla.

La clave primaria se utiliza para garantizar que no haya filas duplicadas en la tabla y que cada fila tenga una identificación única.

En una tabla, la clave primaria se define mediante la selección de una o más columnas que tienen valores únicos en cada fila.

Estas columnas se definen como clave primaria y se utiliza para relacionar datos entre diferentes tablas en una base de datos.

Las claves primarias son importantes porque ayudan a garantizar la integridad de los datos en una base de datos.

Al garantizar que cada fila tenga una identificación única, se evita la posibilidad de datos duplicados o información incompleta.

## **4.- Concepto de clave externa o clave foránea (FK)**

Una **clave externa (FK)** es una columna o varias columnas en una tabla que se relaciona con una columna o varias columnas en otra tabla.

La columna que se relaciona en la otra tabla es la clave principal.

La relación entre las dos tablas se establece cuando se igualan los valores de la clave externa con los valores de la clave principal.

## **6.- Modelo copo de nieve**

El modelo "copo de nieve" es una variante del modelo de datos relacional que se caracteriza por su estructura jerárquica y su capacidad para optimizar el uso del espacio de almacenamiento y mejorar la eficiencia de las consultas.

Esta estructura jerárquica permite reducir la redundancia de datos en la base de datos, lo que significa que se utilizan menos recursos de almacenamiento.

También hace que las consultas sean más eficientes, ya que los datos se almacenan en una estructura más lógica y jerarquizada.

# **Función CALCULATE**:

En Power BI, la función CALCULATE se utiliza para modificar el contexto de evaluación de una fórmula o expresión. 

Permite agregar o eliminar filtros en una **columna calculada** o **medida**, lo que puede cambiar los resultados de las visualizaciones o informes.

La función CALCULATE puede ser utilizada junto con otras funciones de agregación, como **SUM**, AVERAGE, MIN, **MAX**, entre otras, para calcular valores en un contexto específico.

Por ejemplo, se puede utilizar la función CALCULATE para calcular la suma de ventas de un producto sólo para una región específica o para un período de tiempo determinado.

## **Sintaxis:**



```
CALCULATE(<expresión a calcular>, <filtros>)
```



Donde <expresión a calcular> es la medida o columna que se desea calcular y <filtros> son los filtros que se desean aplicar a la expresión.

Los filtros pueden ser especificados utilizando funciones como FILTER, ALL, VALUES, entre otras.



## **Campo Calculado (TABLA: Producto): Precio Unitario**



```
PrecioUnitario = CALCULATE(MAX(VENTA[UnitPrice]))
```



## **Campo Calculado (TABLA: VENTA): Sub-Total**



```
Sub Total = VENTA[Quantity]*VENTA[UnitPrice]*(1-VENTA[Discount])
```



## **Medida: Monto**

Venta = CALCULATE(SUM(VENTA[Sub Total]))
